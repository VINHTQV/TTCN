```python
# Importing library
import folium
import folium.plugins as plugins
import pandas as pd
import json
import requests
import time
from folium.plugins import HeatMapWithTime
import datetime
import numpy as np
```

```python
# Importing Data
df=pd.read_csv('covid.csv', error_bad_lines=False)
df.head(10)
```

```python
# Kiểm tra xem có bất kỳ giá trị null nào trong tập dữ liệu không
df.isnull().sum()
```

```python
# Kiểm tra kiểu dữ liệu
df.dtypes
```

```python
df['Day']= pd.to_datetime(df['Day'])
```

```python
df
```

```python
df['Day'] = df['Day'].apply(lambda x: x.strftime('%Y-%m-%d'))
time_index = []
for i in df['Day'].unique():
    time_index.append(i)
```

```python
time_index
```

```python
lat_long_list = []
for i in df['Day'].unique():
    temp=[]
    for index, instance in df[df['Day'] == i].iterrows():
        temp.append([instance['Latitude'],instance['Longitude']])
    lat_long_list.append(temp)
```

```python
lat_long_list
```

```python
#Choosing the map type 
m = folium.Map(location=[10.7812034,106.655273],zoom_start = 11, tiles="https://tiles.stadiamaps.com/tiles/alidade_smooth_dark/{z}/{x}/{y}{r}.png",attr="Stadia.AlidadeSmoothDark") 
HeatMapWithTime(lat_long_list,radius=5,auto_play=True,position='bottomright',name="cluster",index=time_index,max_opacity=0.7).add_to(m)
m

```

```python
#saving it as a HTML file 
m.save('fo.html')
```
